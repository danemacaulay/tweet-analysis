```python
import spacy
import numpy as np
```

```python
nlp = spacy.load('models/twitter-27B-200')
```

```python
texts = df['text'].tolist()
```

```python
vectors = []
vectors_norm = []
for t in texts:
    doc = nlp(t)
    vectors.append(doc.vector)
    vectors_norm.append(doc.vector_norm)
```

```python
df['vectors'] = list(np.array(vectors))
```

```python

df["likes_normalized"] = df["likes"] / df["likes"].max()
df["replies_normalized"] = df["replies"] / df["replies"].max()
df["retweets_normalized"] = df["retweets"] / df["retweets"].max()
```

```python
features_list = []
for index, row in df.iterrows():
    stats = ([row['likes_normalized']] * 25) + ([row['replies_normalized']] * 25) + ([row['retweets_normalized']] * 25) 
    val = np.append(row['vectors'], stats)
    features_list.append(val)
```

```python
len(features_list[0])
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn import decomposition
```

```python
clusters = kmeans.fit_predict(features_list)
```

```python
clusters[0]
```

```python
df['clusters'] = clusters
```

```python
df_csv = df.drop(['full_vector', 'html','vectors', 'vectors_norm', 'user'], axis=1)
```

```python
df_csv
```

```python
df_csv = df_csv.sort_values(by=['clusters'])
df_csv.to_csv('clustering_analysis.csv')
```

```python
counts = df_csv['clusters'].value_counts()
```

```python
pca = decomposition.PCA(n_components=50)
features_list_50 = pca.fit_transform(features_list)
```

```python
range_n_clusters = list(range(2, 11))
for n_clusters in range_n_clusters:
    print(n_clusters)
    kmeans = KMeans(n_clusters=n_clusters)
    clusters = kmeans.fit_predict(features_list_50)
    silhouette_avg = silhouette_score(features_list_50, clusters)
    print('number of clusters: {}, silhoutte avg: {}'.format(n_clusters, silhouette_avg))
```
